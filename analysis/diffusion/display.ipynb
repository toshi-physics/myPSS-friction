```python
import numpy as np
```

```python
tf = 4000
dt_dump = 100
times = np.arange(0, tf, dt_dump)

mx = 100; my = 100
tol = 0.001
x   = np.linspace(0+tol, 1-tol, mx)
y   = np.linspace(0+tol, 1-tol, my)
xv, yv  = np.meshgrid(x,y)

datadir = '../../data/diffusion/'
rho = np.zeros((tf//dt_dump, mx, my))
i=0
for t in times:
    rho[i] = np.loadtxt(datadir+'rho.csv.'+str(t), delimiter=',')
    i+=1
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

fig, ax=plt.subplots()
#colors = ['#c1272d', '#eecc16','#008176','#0000a7']
#cmap1 = LinearSegmentedColormap.from_list("mycmap", colors)
i=-1
rhomax = np.max(rho[i]); rhomin=np.min(rho[i]); print(rhomax, rhomin)
#ctrf = [ax.pcolormesh(xv, yv, S_sqrd[i], cmap=cmap1, vmin=0, vmax=Smax), ax.quiver(xv,yv,nx[i],ny[i], color='w')]
ctrf = [ax.pcolormesh(xv, yv, rho[i], vmin=0, vmax=rhomax)]
plt.colorbar(ctrf[0])
plt.title(r"$\rho$")
plt.show()
```
